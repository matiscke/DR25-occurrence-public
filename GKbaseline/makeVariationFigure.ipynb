```python
import os
import requests
import pandas as pd
from astropy.io import fits
from cStringIO import StringIO
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
baselineZeta = [0.013,0.009,0.006]
kicZeta = [0.026,0.014,0.010]
sc0p9Zeta = [0.013,0.012,0.006]
noVEZeta = [0.011,0.008,0.005]
noMesSmearZeta = [0.016,0.011,0.007]

xsize = 5
capsize = 8

fig,ax = plt.subplots(figsize=[15, 10])
rect = patches.Rectangle((0,baselineZeta[0]-baselineZeta[2]), 6, 
                         baselineZeta[1]+baselineZeta[2], facecolor = 'r', alpha = 0.2)
plt.plot([0, xsize], [baselineZeta[0], baselineZeta[0]], 'r', alpha = 0.6)
ax.add_patch(rect)

plt.errorbar(1, kicZeta[0], yerr=[[kicZeta[2]],[kicZeta[1]]], capsize = capsize, color = 'k', fmt='--o')
plt.errorbar(2, sc0p9Zeta[0], yerr=[[sc0p9Zeta[2]],[sc0p9Zeta[1]]], capsize = capsize, color = 'k', fmt='--o')
plt.errorbar(3, noVEZeta[0], yerr=[[noVEZeta[2]],[noVEZeta[1]]], capsize = capsize, color = 'k', fmt='--o')
plt.errorbar(4, noMesSmearZeta[0], yerr=[[noMesSmearZeta[2]],[noMesSmearZeta[1]]], capsize = capsize, color = 'k', fmt='--o')

plt.xlim(0,xsize)
plt.ylim(0,0.05)
plt.ylabel("$\zeta_{\oplus}$ with reliability", fontsize = 32)
plt.tick_params(labelsize = 24)
plt.savefig("zetaVariations.pdf",bbox_inches='tight')



```
